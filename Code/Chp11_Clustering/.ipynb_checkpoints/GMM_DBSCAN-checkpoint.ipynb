## GMM: 可处理非球形高斯，但不能处理非高斯形状的簇

```python
import numpy as np

from matplotlib import pyplot as plt

#合成数据
from sklearn import datasets

from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
```

```python
#产生球形高斯分布数据
n_samples = 1500
random_state = 170
X, y =  datasets.make_blobs(n_samples=n_samples, random_state=random_state)

#非球形高斯
transformation = [[0.60834549, -0.63667341], [-0.40887718, 0.85253229]]
X_aniso = np.dot(X, transformation)

#GMM聚类
gmm = GaussianMixture(n_components=3, random_state=random_state)
gmm.fit(X_aniso)
y_pred = gmm.predict(X_aniso)

#显示聚类结果
plt.scatter(X_aniso[:, 0], X_aniso[:, 1], c=y_pred)
plt.title("Anisotropicly Distributed Blobs by GMM")
```

```python
#产生非高斯分布数据
X, y = datasets.make_circles(n_samples=n_samples, factor=.5,
                                      noise=.05)

#GMM聚类
gmm = GaussianMixture(n_components=2, random_state=random_state)
gmm.fit(X)
y_pred = gmm.predict(X)

#显示聚类结果
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.title("Circle Distributed Data by GMM")
```

## DBSCAN可处理非高斯形状的簇

```python
#DBSCAN聚类
from sklearn.cluster import DBSCAN

eps = 0.1
min_samples = 4

db = DBSCAN(eps = eps, min_samples = min_samples)
db.fit(X)
y_pred = db.labels_

#显示聚类结果
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.title("Circle Distributed Data by DBSCAN")
```
