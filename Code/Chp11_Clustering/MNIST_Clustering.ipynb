DBSCAN：DBSCAN
基于密度峰值的聚类：

DBSCAN中MinPts为邻域中邻居的数目，默认是5。

```python
#导入必要的工具包
import pandas as pd
import numpy as np

import time

from matplotlib import cm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#读取训练数据,训练集中有42000个样本
train = pd.read_csv('./data/MNIST_train.csv')

y_train = train.label.values
X_train = train.drop("label",axis=1).values

#将像素值[0,255]  --> [0,1]
X_train = X_train / 255.0

# 原始输入的特征维数和样本数目
print('the shape of train_image: {}'.format(X_train.shape))
```

```python
#显示示例图像
plt.figure(figsize=(14,12))
for digit_num in range(0,64):
    plt.subplot(8,8,digit_num+1)
    grid_data = X_train[digit_num].reshape(28,28)  # reshape from 1d to 2d pixel array
    plt.imshow(grid_data, interpolation = "none", cmap = 'gray' )
    plt.xticks([])
    plt.yticks([])
plt.tight_layout()
```

```python
n_digits = len(np.unique(y_train))

#假设每个数字可细分为20簇
MAX_K = n_digits * 20

random_state = 170
```

```python
# define some utility function for the rest of the notebook
#打印一个数字图像，及其真实标签
def plot_d(digit, label):
    plt.axis('off')
    plt.imshow(digit.reshape((28,28)), cmap=plt.cm.gray)
    plt.title(label)

#打印一组数字图像，25个图像一行
def plot_ds(digits, title, labels):
    n=digits.shape[0]
    n_rows=n/25+1
    n_cols=25
    plt.figure(figsize=(n_cols * 0.9, n_rows * 1.3))
    plt.subplots_adjust(wspace=0, hspace=0)
    plt.suptitle(title)
    for i in range(n):
        plt.subplot(n_rows, n_cols, i + 1)
        plot_d(digits[i,:], "%d" % labels[i])
        
def plot_clusters(y_predict, y, stats, n_clusters):
    #for i in range(n_clusters):
    for i in range(2):
        indices = np.where(y_predict == i)
        title = "Most freq item %d, cluster size %d, majority %d " % (stats[i,2], stats[i,1], stats[i,0])
        
        #打印该簇前25个图像
        plot_ds(X_selected[indices][:25], title, y[indices])

#y_predict:所有样本所属的簇标签
#y：所有样本的真实标签
def clusters_stats(y_predict, y, n_clusters):
    stats = np.zeros((n_clusters,3))
    
    #每个簇
    for i in range(n_clusters):
        indices = np.where(y_predict == i)  #所属该簇的所有样本的索引
        cluster = y[indices]                #这些样本的真实标签
        stats[i,:] = clust_stats(cluster)   #分析本簇样本的真实标签
    return stats

#cluster为某簇所有样本的真实标签
def clust_stats(cluster):
    class_freq = np.zeros(n_digits)
    for i in range(n_digits):
        class_freq[i] = np.count_nonzero(cluster == i)  #该簇中真实标签为10个数字的样本样本数
    
    most_freq = np.argmax(class_freq)  #样本数最多的数字
    n_majority = np.max(class_freq)    #样本数
    n_all = np.sum(class_freq)         #该簇的总样本数
    return (n_majority, n_all, most_freq)

#用每个簇出现最多的数字，作为该簇样本的预测标签，计算平均分类正确率
def clusters_purity(clusters_stats):
    majority_sum  = clusters_stats[:,0].sum()
    n = clusters_stats[:,1].sum()
    return majority_sum / n
```

```python
#4w+样本太多，随机抽取其中20%做实验
from sklearn.model_selection import train_test_split
#from sklearn.utils import shuffle
X_selected, X_val, y_selected, y_val = train_test_split(X_train, y_train, train_size = 0.2,random_state = 0)

# 原始输入的特征维数和样本数目
print('the shape of input: {}'.format(X_selected.shape))
```

```python
#PCA降维，保留85%的能量
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

X_std = StandardScaler().fit_transform(X_selected)
pca = PCA(n_components = 0.85 )

print("PCA begin...")
pca.fit(X_selected)

# 降维 
X_pca = pca.transform(X_std)
print("PCA end.")
print("PCA select %d number of components " %pca.n_components_ )
```

## KMeans: 最佳的K
肘部法

```python
from sklearn.cluster import KMeans

#训练集上的残差平方和
sse = []

for k in range(10, MAX_K, 10):
    kmeans = KMeans(n_clusters=k, random_state=random_state)
    kmeans.fit(X_pca)
    sse.append(kmeans.inertia_)
    
x = range(10, MAX_K, 10)
plt.plot(x, sse)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()
```

这里肘部位置并不明显，取K=50

```python
OPT_K = 50
```

```python
#K均值聚类
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = OPT_K)
y_pred_kmeans = kmeans.fit_predict(X_pca)

stats = clusters_stats(y_pred_kmeans, y_selected, MAX_K)
purity = clusters_purity(stats)
print("Overall purity of KMeans: %f" % purity)

#print("Plotting the first 2 clusters)
plot_clusters(y_pred_kmeans, y_selected, stats, MAX_K)
```

## GMM

```python
#GMM
from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(n_components=OPT_K, random_state=random_state)
gmm.fit(X_pca)
y_pred_gmm = gmm.predict(X_pca)

stats = clusters_stats(y_pred_gmm, y_selected, MAX_K)
purity = clusters_purity(stats)
print("Overall purity of GMM: %f" % purity)

print("Plotting the first 2 clusters")
plot_clusters(y_pred_gmm, y_selected, stats, MAX_K)
```

GMM比K均值聚类效果稍好。

## 层次聚类

```python
#画层次结构图,观测层次聚类的最佳聚类数目
from scipy.cluster.hierarchy import dendrogram
import scipy.cluster.hierarchy as sch

linkage_matrix = sch.linkage(X_pca,  method='ward')
dendrogram(
    linkage_matrix,
    truncate_mode="lastp",  # show only the last p merged clusters
    p=200,  # show only the last p merged clusters
    show_leaf_counts=False,  # numbers in brackets are counts, others idx
    leaf_rotation=60.,
    leaf_font_size=8.,
    show_contracted=True,  # to get a distribution impression in truncated branches
)
plt.title('Hierarchical Clustering Dendrogram')
plt.ylabel("Euclidean Distance")
plt.show()
```

可以看出，在距离接近200左右时，再合并簇会带来较大的距离增长
所以在层次聚类时，我们设置距离阈值为100（0.19版本不支持参数distance_threshold）,簇的数目约为100

```python
K_Ag = 100
```

```python
#层次聚类
from sklearn.cluster import AgglomerativeClustering

hierarchical_cluster = AgglomerativeClustering(n_clusters=K_Ag, linkage='ward')
y_pred_hierarchical = hierarchical_cluster.fit_predict(X_pca)

stats = clusters_stats(y_pred_hierarchical, y_selected, K_Ag)
purity = clusters_purity(stats)
print("Overall purity of Agglomerative Clustering: %f" % purity)

print("Plotting the first 2 clusters")
plot_clusters(y_pred_hierarchical, y_selected, stats, K_Ag)
```

## DBSCAN

```python
#DBSCAN超参数调优，k-distance graph
from sklearn.neighbors import NearestNeighbors

n_samples = len(X_pca)

# 构建空列表，用于保存不同参数组合下的结果
res = []
# min_samples候选值
min_samples_candiates = range(10,35,5)
for j, min_samples in enumerate(min_samples_candiates):
    #min_samples = 15
    nbrs = NearestNeighbors(n_neighbors = min_samples).fit(X_pca)
    distances, _ = nbrs.kneighbors(X_pca)

    distanceDec = sorted(distances[:,min_samples-1])
    res.append(distanceDec)
    plt.plot(list(range(1,n_samples+1)), distanceDec,label = str(min_samples))
```

从k-distance graph图中没有看出distance显著增大的地方，
因此不好确定DBSCAN的超参数eps

不同min_samples对应的k-distance graph形状类似

```python
min_samples = 10
nbrs = NearestNeighbors(n_neighbors = min_samples).fit(X_pca)
distances, _ = nbrs.kneighbors(X_pca)

distanceDec = sorted(distances[:,min_samples-1])
res.append(distanceDec)
plt.plot(list(range(1,n_samples+1)), distanceDec,label = str(min_samples))
```

```python
#距离的变化
delta_distance = np.zeros(n_samples-1)
for i in range(0, n_samples-1, 1):
    #delta_distance[i] = (distanceDec[i+1] - distanceDec[i])/distanceDec[i]
    delta_distance[i] = distanceDec[i+1] - distanceDec[i]
   
#plt.plot(list(range(1,1001)), delta_distance[0:1000])
#plt.plot(list(range(6000,n_samples-1)), delta_distance[6000:])
plt.plot(list(range(1,n_samples)), delta_distance)
```

```python
#DBSCAN
from sklearn.cluster import DBSCAN

# 构建空列表，用于保存不同参数组合下的结果
res = []

# min_samples候选值
min_samples = 10
#min_samples_candiates = range(5,15,2)
eps_indexs = range(500, 6000, 500)
#for j, min_samples in enumerate(min_samples_candiates):
for j, eps_index in enumerate(eps_indexs):
    eps = distanceDec[eps_index]
    dbscan = DBSCAN(eps = eps, min_samples = min_samples)
        
    # 模型拟合
    dbscan.fit(X_pca)
    y_pred_dbscan = dbscan.labels_
        
    # 统计簇的聚类数目（-1表示异常点）
    n_clusters = len([i for i in set(y_pred_dbscan) if i != -1])
        
    # 异常点的数目
    n_outliners = np.sum(np.where(y_pred_dbscan == -1, 1,0))
        
    #核心点的数目
    n_cores = dbscan.components_.shape[0]
        
    # 统计每个簇的样本数目
    stats = str(pd.Series([i for i in y_pred_dbscan if i != -1]).value_counts().values)
    res.append({'eps':eps,'min_samples':min_samples,'n_clusters':n_clusters,'n_cores': n_cores, 'n_outliners':n_outliners,'stats':stats})

# 存储聚类结果       
df = pd.DataFrame(res)
```

可以看出，噪声点太多，绝大部分部分样本都没有被聚类
要使更多的样本成功聚类，需要减少eps,减少min_samples

第一个簇很大，其余簇太小，因为核心点和边界点都优先加入最早出现的簇

```python
#最优参数
eps = distanceDec[780]
dbscan = DBSCAN(eps = eps, min_samples = 10)
        
# 模型拟合
dbscan.fit(X_pca)
y_pred_dbscan = dbscan.labels_

n_clusters = len([i for i in set(y_pred_dbscan) if i != -1])
stats = clusters_stats(y_pred_dbscan, y_selected, n_clusters)
purity = clusters_purity(stats)
print("DBSCAN: %d clusters" %n_clusters)

# 异常点的数目
n_outliners = np.sum(np.where(y_pred_dbscan == -1, 1,0))
print("number of noise: %d " %n_outliners)

print("Overall purity of DBSCAN: %f" % purity)

print("Plotting the first 2 clusters")
plot_clusters(y_pred_dbscan, y_selected, stats, n_clusters)
```
