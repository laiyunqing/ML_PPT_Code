```python
# 绘制回归的损失函数
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#Huber损失的参数
delta = 1.35
sigma = 1.5		

#Huber损失
huber = lambda x: (x**2/2) * (abs(x) <= delta) + (2*delta * abs(x) - delta**2) * (abs(x) > delta)
l_huber = lambda x: sigma + huber(x)*sigma

#L2损失
l2 = lambda x: (x**2)/2

#L1损失
l1 = abs

#画图
funs = [l2, l1, l_huber]
styles = ['r-', 'b:', 'g-.', 'm-']
labels = ['L2', 'L1', 'Huber-'+str(delta)]
x = np.arange(-4, 4, .01)
for i, fun in enumerate(funs):
  plt.plot(x, fun(x), styles[i], label=labels[i])
plt.axis([-4, 4, -0.5, 5])
plt.legend()
plt.show()
```
